Analyse exploratoire du DataFrame pour relever les tendances et visualiser les variables explicatives. 

```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
import seaborn as sns 

df=pd.read_csv('train_data.csv')
df.head()
```

```python
df.shape
```

```python
df.dtypes.value_counts()
```

```python
df.isna().sum()
```

```python
sns.histplot(df['prix'], bins=30, kde=True)  # Ajuste le nombre de bins si nécessaire
plt.xlabel("Prix")
plt.ylabel("Effectif")
plt.title("Distribution des prix")
plt.show()
```

```python
df.describe()
```

```python
for col in df.select_dtypes('float'):
    plt.figure()
    sns.histplot(df[col], kde=True, bins=30, kde_kws={'bw_adjust': 2})  # Ajustement de la bande passante
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Density')
    plt.show()
```

```python
df['surface_totale'] = df[['m2_jardin', 'm2_etage', 'm2_soussol']].sum(axis=1)
df.describe()
```

Target/note

```python
notation = df[['vue_note','etat_note','design_note']]

for col in notation:
    plt.figure()
    sns.boxplot(x=col, y='prix', data=df)
```

```python
df.describe()
```

```python
surfaces = df[['m2_interieur','m2_jardin','m2_etage','m2_soussol']]
#'nb_chambres','nb_sdb','m2_interieur','m2_jardin','m2_etage','m2_soussol','nb_etages'
for col in surfaces:
    plt.figure()
    plt.scatter(x=col, y='prix', data=df)
    plt.xlabel(col)
    plt.ylabel("Prix")
```

```python
historique = df[['annee_construction','annee_renovation']]
for col in historique:
    plt.figure(figsize=(8, 5))
    sns.scatterplot(x=df[col], y=df['prix'], alpha=0.5)
    plt.xlabel(col)
    plt.ylabel("Prix")
    plt.title(f"Relation entre {col} et Prix")
    plt.show()
```

```python
structure = df[['nb_chambres','nb_sdb','nb_etages']]
for col in structure:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=df[col], y=df['prix'])
    plt.xlabel(col)
    plt.ylabel("Prix")
    plt.title(f"Relation entre {col} et Prix")
    plt.show()
```

```python
voisinage = df[['m2_jardin_15voisins','m2_interieur_15voisins']]
for col in voisinage:
    plt.figure(figsize=(8, 5))
    sns.scatterplot(x=df[col], y=df['prix'])
    plt.xlabel(col)
    plt.ylabel("Prix")
    plt.title(f"Relation entre {col} et Prix")
    plt.show()
```
